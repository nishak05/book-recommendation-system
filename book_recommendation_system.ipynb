**we will consider only those user whose has given more than 200 ratings that means teh user who gives ratings more often..and the books only those who got more than 50 rating**

** Collaborative Filteirng Recommender System **

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
Active_User=x[x].index
```

The above will give user id who rated more than 200 books

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(Active_User)]
filtered_rating
```

From all the users we are selecting the info of the active users

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating']
```

So we will get by above is book and no of rating on the book

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

users who has rated minimum 50 books has been selected in famous books..

```python
famous_books
```

now we are checking that these famous books are in filtered book and getting the details

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

This is our final rating..Corresponding to every book which user has rated

```python
final_rating
```

```python
final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

This is the format that is require where user who rated above 200 and book above 50 rating

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

Book1 is represented by 810 dimension say book1 is a vector of 810 dimesnion space..each book is a vector in 810 dimensional space..here we have to find similar vector..we will calculate the similarity score for each vector with every other vectors that will give us the similarity between the two books..

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

we are getting 2D with shape 706 by 706 that means we are getting 706 books similarity with other 706 books

similarity_score

```python
similarity_score[0]
```

This is the first row that tells us that first book 1984 similarity with other books

```python
similarity_score.shape
```

now we will make a function recommend that will take one book and it will suggest 5 books similar to the book

```python
def recom(book_name):
  # here we have to fetch the index of the book that has been passed as
  index=np.where(pt.index==book_name)[0][0]
  #once index is fetched then go that index in similarity score
  similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
  for i in similar_items:
    print(pt.index[i[0]])
```

**The top-5 movies similar to given movies**

```python
recom('1984')
```
